### [警察庁オープンデータ](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/index_opendata.html)で公表されている2019年〜2023年の各本票のcsvファイルを結合して、保存する

各データは以下の通り：   
[２０１９年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2019/opendata_2019.html)   
[２０２０年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2020/opendata_2020.html)   
[２０２１年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2021/opendata_2021.html)   
[２０２２年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2022/opendata_2022.html)   
[２０２３年](https://www.npa.go.jp/publications/statistics/koutsuu/opendata/2023/opendata_2023.html)

```python
import pandas as pd
```

```python
df_2019 = pd.read_csv("honhyo_2019.csv", encoding="shift-jis")
```

```python
df_2020 = pd.read_csv("honhyo_2020.csv", encoding="shift-jis")
```

```python
df_2021 = pd.read_csv("honhyo_2021.csv", encoding="shift-jis")
```

```python
df_2022 = pd.read_csv("honhyo_2022.csv", encoding="shift-jis")
```

```python
df_2023 = pd.read_csv("honhyo_2023.csv", encoding="shift-jis")
```

```python
print(df_2019.columns)
```

```python
print(df_2023.columns)
```

```python
# 2023にあって、2019にないカラムを表示する
print(list(filter(lambda x: x not in df_2019.columns,  df_2023.columns)))
```

```python
# 2019にあって、2023にないカラムを表示する
print(list(filter(lambda x: x not in df_2023.columns,  df_2019.columns)))
```

```python
df_all = pd.concat([df_2019, df_2020, df_2021, df_2022, df_2023], ignore_index=True)
```

```python
df_all
```

```python
# エントリー数を確認する
len(df_2019)+len(df_2020)+len(df_2021)+len(df_2022)+len(df_2023)
```

```python
df_all.info()
```

```python
# 外部結合した全データフレームをCSVファイルとして保存
df_all.to_csv('2019-2023df_edited.csv', index=False)
```

### 全データから「自転車」と「電動アシスト自転車」のデータを抽出して保存する

自転車＝当事者種別「51」   
電動アシスト自転車＝当事者種別「52」

```python
df_all.columns
```

```python
allbike_df = df_all[(df_all['当事者種別（当事者A）'] == 52) | (df_all['当事者種別（当事者B）'] == 52) | (df_all['当事者種別（当事者A）'] == 51) | (df_all['当事者種別（当事者B）'] == 51)]
```

```python
allbike_df.info()
```

```python
# 2019~2023年の過去5年間の事故に絞る
allbike_2019_2023df = allbike_df[allbike_df['発生日時　　年'].isin([2019, 2020, 2021, 2022, 2023])]
```

```python
allbike_2019_2023df.info()
```

```python
# フィルターから外したデータ数の確認
allbike_df[allbike_df['発生日時　　年'].isin([2015, 2016, 2017, 2018])]
```

```python
# 過去５年間のデータフレームをCSVファイルとして保存
allbike_2019_2023df.to_csv('2019-2023自転車合計.csv', index=False)
```

### 単独事故の推移をみる

単独事故＝事故類型「41」

```python
allbike_solo = allbike_2019_2023df[allbike_2019_2023df["事故類型"] == 41]
```

```python
allbike_solo.info()
```

```python
# データフレームをCSVファイルとして保存
allbike_solo.to_csv('単独事故.csv', index=False)
```
